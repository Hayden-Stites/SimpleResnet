```python
import torch
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torchvision.transforms as TT

from torchvision.datasets import ImageFolder
```

```python
data_dir = "./petfinder-adoption-prediction"
resize_dim = 300
total_batches = 292

train_transform = TT.Compose([
    TT.Resize((resize_dim, resize_dim)),
    TT.ToTensor(),
])

dataset = ImageFolder(data_dir+'/train', transform=train_transform)
train_dataloader = DataLoader(dataset=dataset, batch_size=200)
```

```python
def get_mean_and_std(dataloader):
    channels_sum, channels_squared_sum, num_batches = 0, 0, 0
    for data, _ in dataloader:
        channels_sum += torch.mean(data, dim=[0,2,3])
        channels_squared_sum += torch.mean(data**2, dim=[0,2,3])
        num_batches += 1
        print("{:.2f}% complete".format(num_batches * 100/total_batches))
    
    mean = channels_sum / num_batches
    std = (channels_squared_sum / num_batches - mean ** 2) ** 0.5
    return mean, std
```

```python
mean, std = get_mean_and_std(train_dataloader)
```

```python
meanR, meanG, meanB = mean.tolist()
stdR, stdG, stdB = std.tolist()

mean

print("Mean: {:.4f}R, {:.4f}G, {:.4f}B\nSTD: {:.4f}R, {:.4f}G, {:.4f}B"
     .format(meanR, meanG, meanB, stdR, stdG, stdB))
```
